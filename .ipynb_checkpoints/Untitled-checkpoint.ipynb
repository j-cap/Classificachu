```python
def download_pokemon(pokemon, nr=10):
    response = google_images_download.googleimagesdownload()
    arguments = {"keywords":pokemon,"limit":nr,"print_urls":False}
    paths = response.download(arguments)
```

## Rename the files in the download folder

```python
def rename_images(pokemon):
    for count, fname in enumerate(os.listdir(renamePath / pokemon)):
        print(count, fname)
        counter = str(count+1).zfill(4)
        fname_new = pokemon+"_"+counter+".png"
        os.rename(src=renamePath / pokemon /fname, dst=renamePath / pokemon / fname_new)
```

```python
[rename_images(poke) for poke in pokemon[1:]]
```
