```python
n = 5
f, axes = plt.subplots(n,n, figsize=(16,9))
for i in range(n):
    for j in range(n):
        image_data = requests.get(thumbnail_urls[i+n*j])
        image_data.raise_for_status()
        image = Image.open(BytesIO(image_data.content))        
        axes[i][j].imshow(image)
        axes[i][j].axis("off")
plt.show()
```
